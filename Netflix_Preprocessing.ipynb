# Netflix Preprocessing Data
_"Top 5 Netflix Genre in NOV 2019"_


### Setup data
- Import necessary package
- Setup display pandas
- Defines dataframe

```python
import pandas as pd

#read the raw dataset
df = pd.read_csv('netflix_titles_nov_2019.csv')
#set up display based on data set
pd.set_option('display.max_colwidth', 50)
df.head(5)
```

### Filtering
- Filter the collums that will be used
- Filter dataframe by Country = 'United States'
- Filter dataframe by Release_year = '2019'
- Reset the index
- change 'date_added' type data to datetime64[ns]  
- Filter dataframe by month 'November/11'

```python
# df[['col1','col2','col3']]
df_filtered = df[['country','date_added','release_year','rating','listed_in']]

# df.loc[df['col1'] == 'value']
df_country = df_filtered.loc[df_filtered['country'] == 'United States']
df_country.head()
```

```python
# df.loc[df['colhead'] == 'value']
df_year = df_country.loc[df_country['release_year'] == 2019]

# reset_index()
df_year = df_year.reset_index()
df_year.head()
```

```python
# pd.to_datetime(df_year.date_added)
df_year['date_added'] = pd.to_datetime(df_year.date_added)

# [df_year.date_added.dt.month == 11]
df_date = df_year.loc[df_year.date_added.dt.month == 11]
df_date.head()
```

### Export data

```python
# to_csv()
df_date.to_csv('Netflix_filtered.csv')
```
